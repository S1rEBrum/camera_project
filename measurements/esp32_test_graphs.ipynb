```python
import matplotlib.pyplot as plt # type: ignore
```

```python

# DATA FOR 200 BYTES
# dist = [10, 20, 30, 60, 100]
# power_idle = [0.1480, 0.1500, 0.1510, 0.1513, 0.1530]
# power_tx = [0.268, 0.280, 0.36, 0.358, 0.454]
# failures = [0, 3, 11, 25, 78]

# DATA WITH IMAGES
dist = [5, 10, 15, 20, 25, 30]
power_idle = [0.1856, 0.1915, 0.1909, 0.1904, 0.1894, 0.1911]
power_tx = [0.2331, 0.2750, 0.2860, 0.2910, 0.3010, 0.3210]
failures = [0, 1, 3, 2, 4, 3]
times = [2.17, 2.39, 2.66, 2.85, 3.02, 3.64]
```

```python
plt.plot(dist, power_idle, '-o', 0)
plt.grid()
plt.xlabel("Transmission distance [m]")
plt.ylabel("Current [mA]")
plt.title("ESP-NOW Power Consumption in Idle")
ax = plt.gca()
ax.set_xlim([0, 35])
ax.set_ylim([0.175, 0.205])
plt.show()
plt.savefig("test1.svg", format='svg')
```

```python
plt.plot(dist, power_tx, '-o')
plt.grid()
plt.xlabel("Transmission distance [m]")
plt.ylabel("Current [mA]")
plt.title("ESP-NOW Power Consumption in Transmission")
ax = plt.gca()
ax.set_xlim([0, 35])
ax.set_ylim([0.2, 0.35])
plt.show()
```

```python
# plt.plot(dist, power_idle, 'b-o', dist, power_tx, 'g-o')
plt.plot(dist, power_idle, 'b-o',label='power in idle')
plt.plot(dist, power_tx, 'g-o', label="power in transmission")
plt.grid()
plt.xlabel("Transmission distance [m]")
plt.ylabel("Current [mA]")
plt.title("ESP-NOW Power Consumption in Idle VS in Transmission")
plt.legend(loc="upper left")
ax = plt.gca()
ax.set_xlim([0, 35])
ax.set_ylim([0.15, 0.35])
plt.show()
```

```python
plt.bar(dist, failures, width=4.5)
# plt.grid()
plt.xlabel("Transmission distance [m]")
plt.ylabel("Amount of failures")
plt.title("ESP-NOW Failures amount @~250 packets of 200 bytes")
plt.xlim([0, 35])
plt.show()
```

```python
plt.plot(dist, times, '-o')
plt.grid()
plt.xlabel("Transmission distance [m]")
plt.ylabel("Transmission time [s]")
plt.title("ESP-NOW Transmission Time @~250 packets of 200 bytes")
plt.xlim([0, 35])
plt.ylim([1.5, 4])
plt.show()
```
